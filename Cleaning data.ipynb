```python
cn = df.loc[((df['language'] =='ANTLR') |(df['language'] =='Just')| (df['language'] =='Markdown')|
             (df['language'] =='Csound Document')|(df['language'] =='Mermaid')|(df['language'] =='Dockerfile')|
             (df['language'] =='mIRC Script')|(df['language'] =='Nextflow')|(df['language'] =='RMarkdown')|
             (df['language'] =='Nasal')|(df['language'] =='Faust')|(df['language'] =='Jsonnet')|
             (df['language'] =='BitBake')|(df['language'] =='Rich Text Format')|
             (df['language'] =='KiCad Schematic')|(df['language'] =='Procfile')|(df['language'] =='NetLogo')|
             (df['language'] =='Makefile')|(df['language'] =='INI')|(df['language'] =='Asymptote')|
             (df['language'] =='FreeBasic')|(df['language'] =='antler')|(df['language'] =='Nushell'))]

df = df.drop(cn.index)
```

```python
df['language'].value_counts()
```

```python
#Function to clean the description text
def clean_description(description):
    # Remove special characters, emojis, and non-alphanumeric characters
    cleaned_text = re.sub(r'[^\w\s]', '', description)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text.strip()

#Clean the values in the 'description' column
df['description'] = df['description'].apply(clean_description)
```
